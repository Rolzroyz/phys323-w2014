A man whose face is 25 cm away looks into the bowl of a soup spoon and sees his image reflected with a magnification of -0.064. What is the radius of curvature of the spoon? Ans: R = -3:0 cm.


Setup Tips:

1. Recognizing that because the reflected image is smaller ($M < \pm 1$) it will be closer to the mirror (real or imaginary)
2. Convex mirrors _cannot_ produce an inverted image (negitive magnification) so this is a concave mirror
3. Treat the spoon as a semicircler mirror unless explictly told otherwise

```python
from IPython.core.display import Image
Image( filename='homework3-fig1.png')
```

Variables:  
$F$ = focal point  
$H_o$ = object heght | $S_o$ = object distance  
$H_i$ = image heght | $S_i$ = image distance  
$C$ = center | $V$ = vertex  
$R$ = radius | $M$ = Magnification  

Algibraic Solution: 
$S_o = 25cm | M = -0.064$

1. $M = \frac{H_i}{H_o} = -\frac{S_i}{S_o} = -0.064$
2. $- M*S_o = S_i = 1.6cm$
3. $\frac{1}{S_i}+\frac{1}{S_o} = \frac{1}{F} = 1.50 cm^{-1}$
4. $F = -\frac{2}{R} = 0.665 cm$
5. $R = -\frac{2}{F} = -3.0 cm \checkmark$
