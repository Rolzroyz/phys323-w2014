In an interference pattern of the Youngs type, the distance between slits is 0.5 mm, and the wavelength of light is 600nm.

A. If it is desired to have a fringe space of 1mm at the screen what is the proper screen distance?  
B. If a thin plate of glass (n= 1.50) of thickness 100 microns is placed over one of the slits, what is the lateral fringe displacement at the screen?  
C. What path difference corresponds to a shift in the fringe pattern from a peak maximum to the (same) peak half maximum?

Solution
---
__Definitions:__  

- d = slit seperation
- L = screen length
- $\lambda$= wavlength
- FS = fringe spacing
- DS = lateral shift (displacement)
- t = thickness of glass (for part B)
- I<sub>0</sub> = Intensity
- $\Delta$R = path differance
- n = refractive index


```python
1e-3*0.5e-3/(600e-9)
```

__Part (A)__
Using:  
$ FS = \frac{\lambda L}{d}$  
We rearange to get:    
$L = \frac{(FS)d}{\lambda} = \frac{1\times 10^{-3}\bullet 0.5\times 10^{-3}}{600\times10^{9}}=0.833m$

__Part (B)__
The glass adds pathlength to one of the slits laterally displacing the interfernce fringe.
First we find the optical pathlength differance:  
$\frac{nt}{\lambda}-\frac{t}{\lambda}$  
multiply be wavlength to get a physical distance:  
$\lambda\Big(\frac{nt}{\lambda}-\frac{t}{\lambda}\Big)$  
Recognize that that will be eqaul to $d\frac{DS}{L}$  
Write it out and solve for $DS$:
$$ d\frac{(DS)}{L} = \lambda\Big(\frac{nt}{\lambda}-\frac{t}{\lambda}\Big)$$
$$ DS= \frac{(n-1)tL}{d} = \frac{(1.5-1)100\times10^{-6} (0.833)}{0.5\times10^{-3}} = 83.3 mm$$
Devide by the spacing to get the number of peaks:
$$ \frac{83.3mm}{1mm} =83.3 $$

__Part(C)__
Fringe intesity:
$$ I = 4I_0 \cos\Big(\frac{\pi d (FS)}{\lambda L}\Big)$$
Ignoring the constants out in fron we need:
$$ \cos^2 \Big(\frac{\pi d (FS)}{\lambda L}\Big) = 1/2$$
So:
$$ \cos \Big(\frac{\pi d (FS)}{\lambda L}\Big) =\sqrt{1/2}$$
Rembering the 30-60-90 triangle we get:
$$\frac{\pi d (FS)}{\lambda L} = \frac{\pi}{4} = \frac{\Delta R}{\lambda}$$
Where $ \Delta R = \frac{d (FS)}{\lambda}$ (pathlength differance we are looking for)  
Finaly we have:
$$\Delta R = \frac{\lambda}{4} = \frac{600nm}{4} = 150 nm$$

__Summery of Results__  
A 0.833 m  
B 83.3 peaks  
C 150 nm
