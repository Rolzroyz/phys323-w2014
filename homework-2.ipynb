For a prism, with an index of refraction of 1.5, plot $\delta$ vs. $\theta_1$ (see Fig 3-9, p. 61 in your book) for the apex angles of 10 , 20, 30, 40, and 60 degrees. Identify and report $\delta_{min}$ in each case. Make a nice, clear plot from which we can clearly identify the parmeters of each curve.

The equation for the angle of deviation is give by:
$$\delta = \theta_1 +\arcsin{[n \sin{(A - \arcsin{\{\frac{1}{n} \sin{ \theta_1 \}})]}}} - A$$
Where $\delta$ is the angle of deviation, $\theta_1$ is the incident angle, $A$ is the apex angle and $n$ is the index of refraction.
A sketch of the geometry can be seen in figure one.  
The phsyical constraints are such that the light must contact both sides of the prism connected by the Apex angle and the angle $\theta_1$ must be between between 0-90 deg.

```python
#setup, figure 1
import numpy as np
import matplotlib.pyplot as plt
from IPython.core.display import Image
Image( filename='homework2-fig1.png')
```

The next two cells setup and plot the equation above

```python
n = 1.5
theta_min=0
apex_angle = np.array([60.,40.,30.,20.,10.]) #degrees
apex_angler = apex_angle*np.pi/180. #radians
theta = arange(theta_min,90,.1) # simple base axis
tr = theta*np.pi/180.
#transpose and compute the arrays at once for brevity
deviation = np.transpose([tr+np.arcsin(n*np.sin(ar-np.arcsin(np.sin(tr)/n)))-ar for ar in apex_angler])
angle_of_deviation = deviation*180/np.pi
```

```python
fig = plt.figure(figsize=(8,4.5),dpi='200')
ax = gca() #get current axes
ax.plot(theta,angle_of_deviation)
ax.plot([49,31,23,15,8],[37.18,21.73,15.68,10.19,5.024],'k*',markersize=15,linewidth=5)
#label data
ax.set_xlabel(r'Angle of Incidence $\theta_1$ (deg)', fontsize='16')
ax.set_ylabel(r'Angle of Deflection $\delta$ (deg)', fontsize='16')
ax.legend(('$60^o$','$40^o$','$30^o$','$20^o$','$10^o$','$\delta_{min}$'),title='Apex Angle',loc="upper left")

#presentation/as1thetics
ax.grid(True)
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_color('none')
ax.spines['left'].set_visible(False)
ax.spines['right'].set_visible(False)

#Title added last to indicate end of plot
ax.set_title('Deflection Through Prism With n=1.5', fontsize='22', loc='center')

plt.show()
```
