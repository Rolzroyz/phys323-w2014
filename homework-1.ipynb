The goal for this problem is to minimize the time it takes for the light ray to travel from point A to point B. Key assumptions are as follows:

1. The Light must contact point A Point B and the reflected surface
2. All light is reflected perfectly (no interface effects)
3. The medium in homogenous (speed of light does not change)
4. The shortest distance between to points is a strait line

The time $T$ it takes for the particle to get from A to B can be given by the following equation: $$ T = \frac{\sqrt{a^2+x^2}}{V} + \frac{\sqrt{b^2 + (d-x)^2}}{V} $$
Deviding through by constant speed $V$ gives us the formula for the pathlength $P$:
$$P = \sqrt{a^2+x^2} + \sqrt{b^2 + (d-x)^2} $$
The minima may be found by setting the derivative of $P$ with respect to $x$ equal to zero
$$ \frac{2x}{2 \sqrt{a^2+x^2}} + \frac{-2(d-x)}{2 \sqrt{b^2 + (d-x)^2}} = 0$$
This can be simplified to:
$$ \frac{x}{\sqrt{a^2+x^2}} = \frac{d-x}{\sqrt{b^2 + (d-x)^2}}$$
Using the pathagorian theorem and trigonometric relationships in figure one we see that:  
$\sin{\theta_i} = \frac{x}{\sqrt{a^2+x^2}} $ and $sin{\theta_r} = \frac{d-x}{\sqrt{b^2 + (d-x)^2}}$

Substituting this into the equation above we gives that $\sin{\theta_i} = \sin{\theta_r} $ which shows that $\theta_i=\theta_r$ which is the the law of reflection
