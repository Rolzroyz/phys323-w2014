A thin lens of with f=+10 cm is separated by 5 cm from a thin lens with f=-10 cm. Find the equivalent focal point of this combination of lenses. Ans: 20 cm. Hint: the focal length is -1/C in your system matrix (see Eq. 18-25 in your book).

```python
from IPython.core.display import Image
Image( filename='homework5-fig1.png',width = 400,height=150)
```

This is relitivly straitforward to solve using the matix method, start by writing out the matricies for each of the elements:  
Lens A = $\bigl(\begin{smallmatrix}
1&0\\ \frac{-1}{f_A}&1 
\end{smallmatrix} \bigr)$ = $\bigl(\begin{smallmatrix}
1&0\\ \frac{-1}{10} cm^{-1}& 1 
\end{smallmatrix} \bigr)$

Translation = $\bigl(\begin{smallmatrix}
1&L\\ 0&1 
\end{smallmatrix} \bigr)$ = $\bigl(\begin{smallmatrix}
1&5cm\\ 0&1 
\end{smallmatrix} \bigr)$

Lens B = $\bigl(\begin{smallmatrix}
1&0\\ \frac{-1}{f_B}&1 
\end{smallmatrix} \bigr)$ = $\bigl(\begin{smallmatrix}
1&0\\ \frac{1}{10} cm^{-1}& 1 
\end{smallmatrix} \bigr)$

Then multiply them out starting with the rightmost element in figure 1 and moving left to get the total system matix (all units in $cm$, $cm^-1$):

System = $\bigl(\begin{smallmatrix}
1&0\\ \frac{1}{10} & 1 
\end{smallmatrix} \bigr)  \bigl(\begin{smallmatrix}
1&5\\ 0&1 
\end{smallmatrix} \bigr)  \bigl(\begin{smallmatrix}
1&0\\ \frac{-1}{10} & 1 
\end{smallmatrix} \bigr) $ = $\bigl(\begin{smallmatrix}
\frac{1}{2}&5\\ \frac{-1}{20}& \frac{3}{2} 
\end{smallmatrix} \bigr)$  

Remembering that if the system matix is writen int the form of $\bigl(\begin{smallmatrix}
A&B\\ C&D 
\end{smallmatrix} \bigr)$ then $C = \frac{-1}{F}$ where $F$ is the effective focal length we get that:  
$F =\frac{-1}{C}=20cm \checkmark$
