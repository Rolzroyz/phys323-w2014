Using a typical diapole density for glass and a cutoff wavelength inside the ultraviolet spectrum we chose:  
$N = 8.1 \times 10^{28}$ dipoles/m<sup>3</sup>  
$\omega_{0} = 1.6 \times 10^{16}$ Hz (rads/m)  
($\lambda_0 = 111$ nm)

The MKS units for other constants of interest:  
$q = 1.602 \times 10^{-19}$ Coulombs  
$\epsilon_{0} = 8.854 \times 10^{-12}$ Farads/m  
$m_e = 9.109 \times 10^{-31}$ kg  
$c = 2.998 \times 10^{8}$ m/s  


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Material parameters
N = 8.1*10**28     #Dipole density
lo = 111*10**-9    #cutoff wavelength

#desiered plot spectrum
lmin = 85*10**-9   #min wavelength (meters)
lmax = 600*10**-9  #max wavelength (meters)
steps = 5000.      #plot resoltion (points)

#fundamental constants
q = 1.602*10**-19  #fundamental charge
En = 8.854*10**-12 #permiativity of free space
me = 9.109*10**-31 #electron mass
c = 3.0*10**8      #speed of light

#compute w and n
l = np.arange(lmin,lmax+(lmax-lmin)/steps,(lmax-lmin)/steps)
w = 2*np.pi*c/l    #Observed Frequncies
wn = 2*np.pi*c/lo  #Cutoff Frequency
nsqrd = 1 + (N*q**2)/En/me/(wn**2-w**2) 
maskneg = (nsqrd<0)   #find all negitive indexies
n = abs(nsqrd)**.5 #take the squareroot of all values
n[maskneg] = -n[maskneg] # flip the negitive indexies back
```

```python
fig = plt.figure(figsize=(8,4),dpi='200')
ax = fig.add_subplot(111)
ax.plot(w,n,'r.',w[maskneg],n[maskneg],'b.',(wn,wn),(-5,5),'k')

#label data
ax.set_xlabel('Frequancy $\omega$ (rads/sec)', fontsize='14')
ax.set_ylabel('Refactive index $n$', fontsize='14')
ax.legend(('Real', 'Imaginary','$\omega_0$'),loc="lower left")

#Set range
ax.set_ylim(-5, 5)

#presentation/asthetics
ax.grid(True)
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_color('none')
ax.spines['left'].set_color('none')
ax.spines['right'].set_color('none')


#Title added last to indicate end of plot
title('Figure 1: Dispersion Through Lenslike Material', fontsize='18', loc='center')

plt.show()
```

In figure one we see that for $\omega < \omega_0$ the index of refraction are real. This corisponds to the normally dispersive case where the dipoles inside the material are occsillating. As discussed in class at higher frequancies the diapoles have a harder time keeping up with the electric field.  
In the case of the limit we have $\lim_{\omega \to \omega_0^\mp} n \to \pm\infty$ this is somewhat artificial as nothing particularly dramatic happens at $\omega_0$. Other effects such as damping will take over close to the cutoff frequancy.  
For $\omega > \omega_0$ we have that the anomluosly dispersive region where the index of refration increases with frequancy. Note that, the index of refration is purely imaginary here.
