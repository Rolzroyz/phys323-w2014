The Principle Rays:
All originate from the tip of the object. 

Ray A: Travels parallel to the optical axis then refracts out as if originating originating from the leftmost focal point $F$ (which is negative) 
Ray B: Travels towards the positive focal point (also $F$) then refracts parallel to the optical axis  
Ray C: Travels refracted through the vertex $V$

Measured image height = 1.2 cm | $H'_m$ - measured height of the image above the optical axis (positive indicate vertical)
Measured image distance = -8.4 cm | $S'_m$ - measured distance from $I$ to $V$ or (negative indicates imaginary)  

Mathematic Solution:

- $S = 15cm$ | $S$ - distance from Object $O$ to vertex $V$ (positive indicates real image)
- $H = 15cm$ | $H$ - Height of the object
2. $F = -20cm$ | $F$ - focal length of lens (negative indicates divergent)
3. $\frac {1}{S} + \frac{1}{S'} = \frac{1}{F}$ | lens equation
4. $S' = [\frac{1}{F} - \frac{1}{S}]^{-1} = -8.57 cm  \checkmark$ | Expected Image distance  
5. $M = -\frac{S'}{S} = 0.571$ | $M$ - magnification (positive magnification indicates upright image)
6. $H' = M*H = 1.14 cm  \checkmark$

Comparison:

If we compare the measured values to the numeric solutions we get $\frac{H'_m}{H'} = 1.05$ and $\frac{S'_m}{S'} = 0.98$. This shows that they agree within 5% which is better than I might have guessed given my poor drawing skills.
